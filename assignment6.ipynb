```python
import pandas as pd
import numpy as np
import yfinance as yf

df = yf.download("GOOG", start = "2023-01-01", end = "2024-12-31", auto_adjust=True)
df.head()

def SMA(data, window, column = 'Close'):
  return data[column].rolling(window = window).mean()

df['SMA_short'] = SMA(df, 20)
df['SMA_long'] = SMA(df, 50)

df['Signal'] = np.where(df['SMA_short'] > df['SMA_long'], 1, 0)
df['Position'] = df['Signal'].diff()

df['Buy'] = np.where(df['Position'] == 1, df['Close'].squeeze(), np.nan)
df['Sell'] = np.where(df['Position'] == -1, df['Close'].squeeze(), np.nan)

display(df)

position = 0
buy_price = 0
profits = []

for i in range(len(df)):
  position_signal = df['Position'].iloc[i]
  close_price = df['Close'].iloc[i].item()

  if position_signal == 1 and position == 0:
    position = 1
    buy_price = close_price
    print(f'Bought on {df.index[i].date()} for ${buy_price:.2f}(USD)')

  elif position_signal == -1 and position == 1:
    sell_price = close_price
    profit = sell_price - buy_price
    profits.append(profit)
    position = 0
    print(f'Sold on {df.index[i].date()} for ${sell_price:.2f}(USD). Profit: ${profit:.2f}(USD)')

total_profit = sum(profits)
print(f'Total_profit: ${total_profit:.2f}(USD)')
```
